```r
# Sample data (replace with your own data)
sample_scores <- c(65, 72, 68, 75, 71, 70, 73, 69, 74, 68)
```

```r
# Calculate sample statistics
sample_mean <- mean(sample_scores)
sample_sd <- sd(sample_scores)
sample_size <- length(sample_scores)

# Set confidence level
confidence_level <- 0.95
```

```r
# Calculate margin of error
margin_of_error <- qt((confidence_level + 1) / 2, df = sample_size - 1) * (sample_sd / sqrt(sample_size))

# Calculate confidence interval
confidence_interval <- c(sample_mean - margin_of_error, sample_mean + margin_of_error)
```

```r
# Display the results
cat("Sample Mean:", sample_mean, "\n")
cat("Margin of Error:", margin_of_error, "\n")
cat("95% Confidence Interval:", confidence_interval, "\n")
```

Confidence Interval for Population Proportion (z-Distribution)

$$ MOE = z_{\alpha/2} \times \sqrt{\frac{p \cdot (1 - p)}{n}} $$


```r
survey_results <- c(rep(1, 35), rep(0, 15))  # 1 for preference, 0 for no preference

# Calculate sample statistics
sample_proportion <- sum(survey_results) / length(survey_results)
sample_size <- length(survey_results)

# Set confidence level
confidence_level <- 0.95

# Calculate margin of error
margin_of_error <- qnorm((confidence_level + 1) / 2) * sqrt((sample_proportion * (1 - sample_proportion)) / sample_size)

# Calculate confidence interval
confidence_interval <- c(sample_proportion - margin_of_error, sample_proportion + margin_of_error)

# Display the results
cat("Sample Proportion:", sample_proportion, "\n")
cat("Margin of Error:", margin_of_error, "\n")
cat("95% Confidence Interval:", confidence_interval, "\n")
```

Confidence Interval for the Difference in Means (Independent Samples)

$$ MOE = t_{\alpha/2, \text{df}} \times \sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}} $$


```r

group1 <- c(25, 28, 30, 32, 29)
group2 <- c(22, 26, 28, 30, 27)

# Calculate sample statistics
mean_group1 <- mean(group1)
mean_group2 <- mean(group2)
sd_group1 <- sd(group1)
sd_group2 <- sd(group2)
n1 <- length(group1)
n2 <- length(group2)

# Set confidence level
confidence_level <- 0.99

# Calculate margin of error
margin_of_error <- qt((confidence_level + 1) / 2, df = min(n1 - 1, n2 - 1)) * sqrt((sd_group1^2 / n1) + (sd_group2^2 / n2))

# Calculate confidence interval
confidence_interval <- c((mean_group1 - mean_group2) - margin_of_error, (mean_group1 - mean_group2) + margin_of_error)

# Display the results
cat("Difference in Means:", mean_group1 - mean_group2, "\n")
cat("Margin of Error:", margin_of_error, "\n")
cat("99% Confidence Interval:", confidence_interval, "\n")
```
